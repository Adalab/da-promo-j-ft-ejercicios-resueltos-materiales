Requisitos que deben cumplir las gráficas (en caso de usar alguna de ellas):

Todas deben llevar título.

Los nombres de los ejes deben estar en castellano.

En los scatterplot los puntos deben tener un tamaño de 2.

En los scatterplot los puntos deben tener un color diferente al azul.

En los barplot las barras siempre tienen que estar ordenadas de mayor a menor.

En los barplot y countplot las barras tienen que tener un color diferente a los que vienen por defecto.

En los pieplot siempre tenemos que añadir la etiqueta de a que % corresponde cada quesito.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv(r"C:\Users\karla\Desktop\soluciones_alumnas\da-promo-j-ft-ejercicios-resueltos-materiales\Mod_3\world_data_full_apply_nulos.csv", index_col=0)
df.head()
```

```python
# 1.  ¿Cómo se distribuye la tasa de natalidad en el conjunto de datos? ¿Hay algún patrón evidente?
sns.histplot(x="birth_rate", data=df)
```

```python
# Configurar la fuente y la paleta
plt.rcParams['font.family'] = 'Calibri'  # Cambiar la fuente a Calibri
sns.set_palette("inferno")  # Usar la paleta inferno

# Crear la figura y el eje, y establecer el color de fondo a negro
fig, ax = plt.subplots()
fig.patch.set_facecolor('black')  # Cambiar el color de fondo de la figura a negro
ax.set_facecolor('black')         # Cambiar el color de fondo del área de la gráfica a negro

# Crear el histograma
sns.histplot(x="birth_rate", data=df, edgecolor='fuchsia', ax=ax)  # Cambiar el borde de las barras a rosa fucsia

# Ajustes de etiquetas con color y tamaño de fuente personalizados
ax.set_xlabel("Tasa de natalidad", fontsize=16, color='white')  # Cambiar el color y tamaño de la etiqueta del eje X a blanco
ax.set_ylabel("Frecuencia", fontsize=16, color='white')         # Cambiar el color y tamaño de la etiqueta del eje Y a blanco

# Personalizar los ejes
ax.spines['top'].set_visible(False)     # Ocultar línea superior
ax.spines['right'].set_visible(False)   # Ocultar línea derecha
ax.spines['left'].set_linewidth(2.5)    # Hacer la línea del eje Y más gruesa
ax.spines['bottom'].set_linewidth(2.5)  # Hacer la línea del eje X más gruesa
ax.spines['left'].set_color('yellow')   # Cambiar el color de la línea del eje Y
ax.spines['bottom'].set_color('yellow') # Cambiar el color de la línea del eje X

# Cambiar el color de los números de los ejes a amarillo
ax.tick_params(axis='x', colors='yellow')  # Cambiar el color de los números del eje X
ax.tick_params(axis='y', colors='yellow')  # Cambiar el color de los números del eje Y

# Mostrar la gráfica con fondo negro
plt.show()

# Si deseas guardar la imagen con fondo negro
# fig.savefig('histograma_fondo_negro.png')

```

```python
# 2.  ¿Existe una relación aparente entre la tasa de mortalidad infantil y la esperanza de vida? Proporciona una breve descripción de los patrones que observes.
sns.scatterplot(x = 'infant_mortality', 
                y = "life_expectancy", 
                size = 2, 
                color = "black",
                data = df )
plt.ylabel("esperanza de vida")
plt.xlabel("mortalidad infantil");
```

```python
# Establecer el estilo general de la gráfica
sns.set(style="darkgrid")

# Configurar la fuente para la gráfica
plt.rcParams['font.family'] = 'Calibri'  # Cambiar la fuente a Calibri

# Crear la figura y los ejes con un tamaño más grande para presentaciones
fig, ax = plt.subplots(figsize=(8, 4))
fig.patch.set_facecolor('black')  # Cambiar el fondo de toda la figura a negro

# Crear el scatterplot con una paleta de colores cálida para destacar los puntos
scatter = sns.scatterplot(
    x="infant_mortality",
    y="life_expectancy",
    size="infant_mortality",            # Tamaño del punto basado en la tasa de mortalidad infantil
    sizes=(20, 200),                    # Tamaño de los puntos más amplio
    hue="infant_mortality",             # Color del punto basado en la tasa de mortalidad infantil
    palette="magma",                    # Usar la paleta 'magma' para el color
    alpha=0.7,                          # Hacer los puntos semitransparentes
    edgecolor="fuchsia",                # Cambiar el color del borde de los puntos a amarillo
    linewidth=0.8,                      # Espesor del borde de los puntos
    data=df,
    ax=ax,                              # Especificar el eje
)

# Agregar una línea de regresión suavizada con Seaborn para mostrar la tendencia
sns.regplot(
    x='infant_mortality', 
    y='life_expectancy', 
    scatter=False, 
    color='yellow', 
    line_kws={"linewidth": 2, 'alpha': 0.8},
    data=df,
    ax=ax
)

# Ajustar las etiquetas de los ejes con Calibri, tamaño 13, negrita y color negro
ax.set_xlabel("Mortalidad Infantil (por cada 1000 nacidos vivos)", fontsize=13, color='black', weight='bold')
ax.set_ylabel("Esperanza de Vida (en años)", fontsize=13, color='black', weight='bold')

# Cambiar el fondo de los ejes para hacerlo negro
ax.set_facecolor('black')  # Fondo del área de la gráfica a negro
ax.spines['top'].set_visible(False)     # Ocultar línea superior
ax.spines['right'].set_visible(False)   # Ocultar línea derecha
ax.spines['left'].set_linewidth(2.5)    # Hacer la línea del eje Y más gruesa
ax.spines['bottom'].set_linewidth(2.5)  # Hacer la línea del eje X más gruesa
ax.spines['left'].set_color('yellow')   # Cambiar el color de la línea del eje Y a amarillo
ax.spines['bottom'].set_color('yellow') # Cambiar el color de la línea del eje X a amarillo

# Cambiar el color de los números de los ejes a amarillo
ax.tick_params(axis='x', colors='yellow', labelsize=12)  # Cambiar el color y tamaño de los números del eje X a amarillo
ax.tick_params(axis='y', colors='yellow', labelsize=12)  # Cambiar el color y tamaño de los números del eje Y a amarillo
'fuchsia'
# Personalizar la leyenda
legend = ax.legend(title='Mortalidad Infantil', fontsize=12, title_fontsize='13', loc='upper right', frameon=True, facecolor='black', edgecolor='yellow')
plt.setp(legend.get_texts(), color='white')  # Cambiar el color del texto de la leyenda a blanco
plt.setp(legend.get_title(), color='white')  # Cambiar el color del título de la leyenda a blanco

# Título para la gráfica
plt.title('Relación entre la Mortalidad Infantil y la Esperanza de Vida', fontsize=14, color='white', pad=35)

# Mostrar la gráfica con fondo negro
plt.show()
```

```python
# 3.  ¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos? Proporciona el porcentaje de países en ese continente.
df_pais_continente = df.groupby("continent")["country"].count().reset_index()
df_pais_continente.rename(columns = {"country": "%"}, inplace = True)
df_pais_continente["%"] = np.round(df_pais_continente["%"] * 100 / df_pais_continente["%"].sum(), 2)
df_pais_continente.sort_values(by = "%", ascending = False, inplace=True)
df_pais_continente
```

```python
# grafica 1
sns.barplot(x="continent", y="%", data=df_pais_continente)
plt.xticks(rotation=90)
plt.xlabel("")
```

```python
# Lista de colores especificados para las barras y las líneas de la cuadrícula
colores = ['red', 'orange', 'yellow', 'green', 'blue', 'pink', 'violet']

# Crear la figura y los ejes con un tamaño más grande para presentaciones
fig, ax = plt.subplots(figsize=(7, 4))
fig.patch.set_facecolor('black')  # Cambiar el fondo de toda la figura a negro
ax.set_facecolor('black')         # Cambiar el fondo del área de los ejes a negro

# Crear la gráfica de barras con los colores personalizados
sns.barplot(x="continent", y="%", data=df_pais_continente, palette=colores, ax=ax)

# Ajustes adicionales para rotar las etiquetas del eje X
plt.xticks(rotation=90)

# Cambiar el color de las etiquetas del eje X usando la misma escala de colores
for i, label in enumerate(ax.get_xticklabels()):
    label.set_color(colores[i])
    label.set_fontsize(12)
    label.set_weight('bold')

# Cambiar el color de los números del eje Y a amarillo
for tick in ax.get_yticklabels():
    tick.set_color('yellow')
    tick.set_fontsize(12)

# Personalizar los ejes (spines) para que sean amarillos
ax.spines['top'].set_visible(False)     # Ocultar la línea superior
ax.spines['right'].set_visible(False)   # Ocultar la línea derecha
ax.spines['left'].set_linewidth(2.5)    # Hacer más gruesa la línea del eje Y
ax.spines['bottom'].set_linewidth(2.5)  # Hacer más gruesa la línea del eje X
ax.spines['left'].set_color('yellow')   # Cambiar el color de la línea del eje Y a amarillo
ax.spines['bottom'].set_color('yellow') # Cambiar el color de la línea del eje X a amarillo

# Cambiar las líneas de la cuadrícula para que coincidan con los colores de las barras
ax.grid(True, axis='y', linestyle='--', linewidth=1.5)
gridlines = ax.get_ygridlines()
for i, line in enumerate(gridlines):
    if i < len(colores):  # Asegurarse de no exceder la cantidad de colores
        line.set_color(colores[i])
        line.set_alpha(0.6)  # Hacer la línea un poco transparente para que no distraiga demasiado

# Etiqueta del eje Y con ajustes de color y tamaño
plt.ylabel('%', fontsize=14, color='yellow', weight='bold')

# Mostrar la gráfica con fondo negro
plt.show()

```

```python
# grafica 2
colores = ["c", "cadetblue", "mediumseagreen", "seagreen", "teal", "turquoise", "lightseagreen"]

plt.pie(df_pais_continente["%"], 
        labels = df_pais_continente["continent"], 
        autopct=  '%1.1f%%', 
        textprops={'fontsize': 6}, 
        colors= colores);
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Datos simulados para los porcentajes
continentes = ["Africa", "Europe", "Asia", "Central America", "Oceania", "South America", "North America"]
porcentajes = [27.3, 24.2, 23.2, 10.3, 7.2, 6.7, 1.0]

# Usar la paleta inferno de matplotlib para todos los continentes
colores = [cm.inferno(i / len(continentes)) for i in range(len(continentes))]

# Crear la figura con un tamaño adecuado
fig, ax = plt.subplots(figsize=(5, 8), facecolor='black')

# Crear el gráfico de pastel (anillo)
wedges = ax.pie(
    porcentajes, 
    labels=None,  # Eliminar etiquetas del gráfico
    autopct=None, # No mostrar porcentajes en el gráfico
    colors=colores, 
    startangle=90, 
    wedgeprops={'edgecolor': 'white', 'linewidth': 0.8},  # Bordes blancos y más delgados
    explode=[0.05] * len(continentes)  # Separar todas las secciones para resaltarlas
)[0]  # Solo necesitamos los wedges para la leyenda

# Añadir un círculo en el centro para dar el efecto de "doughnut"
centre_circle = plt.Circle((0, 0), 0.70, fc='black')
fig.gca().add_artist(centre_circle)

# Añadir un título al gráfico centrado
plt.title('Distribución de Continentes (%)', color='white', fontsize=16, weight='bold', pad=20, loc='center')

# Añadir una leyenda para explicar las categorías, ubicada abajo a la derecha
ax.legend(
    wedges, [f"{continente} - {porcentaje:.1f}%" for continente, porcentaje in zip(continentes, porcentajes)],
    title="Continentes",
    loc="lower right",
    fontsize=12,
    title_fontsize='13',
    frameon=True,
    facecolor='black',  # Fondo de la leyenda en negro
    edgecolor='white'   # Borde de la leyenda en blanco
)
plt.setp(ax.get_legend().get_texts(), color='white')  # Cambiar el color del texto de la leyenda a blanco
plt.setp(ax.get_legend().get_title(), color='white')  # Cambiar el color del título de la leyenda a blanco

# Ajustar la visualización para asegurar que todo se ve adecuadamente
plt.tight_layout()

# Mostrar la gráfica
plt.show()

```

```python
# 5.  ¿Existen relaciones visibles entre la tasa de natalidad y las siguientes variables: tasa de mortalidad infantil, tasa de impuestos total, densidad de población y tasa de alfabetización?
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(40, 20))

sns.scatterplot(x="birth_rate", y="infant_mortality", data=df, ax=axes[0][0])
sns.scatterplot(x="birth_rate", y="life_expectancy", data=df, ax=axes[0][1])
sns.scatterplot(x="birth_rate", y="density", data=df, ax=axes[1][0])
sns.scatterplot(x="birth_rate", y="urban_population", data=df, ax=axes[1][1])

# ponemos titulos a las gráficas
axes[0][0].set_title("tasa de natalidad vs mortalidad infantil")
axes[0][1].set_title("tasa de natalidad vs esperanza de vida")
axes[1][0].set_title("tasa de natalidad vs densidad")
axes[1][1].set_title("tasa de natalidad vs población urbana")


# cambiamos nombres de los ejes
axes[0][0].set_xlabel("Tasa de natalidad")
axes[0][1].set_xlabel("Tasa de natalidad")
axes[1][0].set_xlabel("Tasa de natalidad")
axes[1][1].set_xlabel("Tasa de natalidad")

axes[0][0].set_ylabel("Mortalidad infantil")
axes[0][1].set_ylabel("Esperanza de vida")
axes[1][0].set_ylabel("Densidad")
axes[1][1].set_ylabel("Poblacion urbana")
```
